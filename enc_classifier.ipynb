```python
import spacy                        
nlp = spacy.load('en',parser=False)
```

```python
def process_encounter(enc_id,enc):
    try:
        doc = nlp(unicode(enc))
        data=[]
        for token in doc:
            if token.is_stop or token.is_punct:
                continue
            else:
                data.append(token.lemma_.lower())
        return [enc_id,' '.join(data)]
    except:
        return [enc_id,'']
import pandas as pd
import glob
files = glob.glob('/Users/cbranzan/personal/enc1000/*')
labels = {}
freq={}
result = []
for file_name in files:
    with open(file_name,'r') as f:
        content =' '.join(f.readlines())
        if len(content)<100:
            continue
#         print 'process ',file_name
        annotations = process_encounter(file_name,content)
        if len(annotations[1])<100:
            continue
        if 'CCU' in content:
            labels[file_name]=1
            if 1 in freq:
                freq[1]+=1
            else:
                freq[1]=1
        elif 'NICU' in content:
                labels[file_name]=2
                if 2 in freq:
                    freq[2]+=1
                else:
                    freq[2]=1
        else:
            labels[file_name]=0
            if 0 in freq:
                freq[0]+=1
            else:
                freq[0]=1
        annotations.append(labels[file_name])
        result.append(annotations)
print freq

new_encs = pd.DataFrame(data=result, columns =['encounter_id','content','label'])
new_encs
```

```python
from sklearn.feature_extraction.text import HashingVectorizer
from sklearn.feature_selection import SelectKBest, chi2
from sklearn.linear_model import Perceptron
from sklearn import metrics
from time import time

train_size=700 #out of 1000
x_train = new_encs[:train_size]['content']
y_train = new_encs[:train_size]['label']
x_test = new_encs[train_size:]['content']
y_test = new_encs[train_size:]['label']

vectorizer = HashingVectorizer(stop_words='english', non_negative=True, n_features=10000)
X_train = vectorizer.transform(x_train)
X_test = vectorizer.transform(x_test)

ch2 = SelectKBest(chi2, k=200)
X_train = ch2.fit_transform(X_train, y_train)
X_test = ch2.transform(X_test)
```

```python
import tensorflow.contrib.learn as learn
from sklearn import datasets, metrics
feature_columns = learn.infer_real_valued_columns_from_input(X_train)
print feature_columns
!rm -r /tmp/enc_classifier/
classifier = learn.DNNClassifier(hidden_units=[100, 200, 100,100,200,100,100,200,100], n_classes=3,feature_columns=feature_columns,model_dir='/tmp/enc_classifier/')

classifier.fit(X_train.toarray(), y_train.values, steps=6000, batch_size=32)
print '====================='
print 'CLASSIFICATION RESULTS'
print 'ON TEST SET:'
preds = list(classifier.predict(X_test.toarray(), as_iterable=True))
score = metrics.accuracy_score(y_test, preds)

print("Accuracy: %f" % score)
print 'CLASSIFICATION REPORT'
print metrics.classification_report(y_pred=preds,y_true=y_test)
print '---------------------'
```
